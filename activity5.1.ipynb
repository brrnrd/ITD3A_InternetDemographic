```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
internet = pd.read_excel("C:\\Users\\User\\Desktop\\ITD3A\\dataset\\internet.xlsx")
```

```python
print(internet.head())
```

<h1>INSIGHT 1</h1>

```python
plt.figure(figsize=(10, 6))
plt.hist(internet['Birth rate'], bins=25, color='lightblue', edgecolor='black')
plt.title('Birth Rate Distribution')
plt.xlabel('Birth Rate')
plt.ylabel('Count')
plt.grid(axis='y')
plt.show()
```

<h1>INSIGHT 2</h1>

```python
top_countries_internet = internet.nlargest(5, 'Internet users')[['Country Name', 'Internet users']]
print("Top 5 Countries with the Most Internet Users:")
print(top_countries_internet)
```

<h1>INSIGHT 3</h1>

```python
income_distribution = internet['Income Group'].value_counts()
plt.figure(figsize=(10, 6))
income_distribution.plot(kind='bar', color='lightcoral')
plt.title('Number of Countries by Income Group')
plt.xlabel('Income Group')
plt.ylabel('Country Count')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

<h1>INSIGHT 4</h1>

```python
plt.figure(figsize=(10, 6))
sns.boxplot(y=internet['Internet users'])
plt.title('Box Plot of Internet Users')
plt.grid(axis='y')
plt.show()
```

<h1>INSIGHT 5</h1>

```python
countries_no_internet = internet[internet['Internet users'] == 0]['Country Name']
print("Countries with No Internet Access:")
print(countries_no_internet)
```

<h1>INSIGHT 6</h1>

```python
avg_birth_rate = internet['Birth rate'].mean()
print(f"Average Birth Rate: {avg_birth_rate:.2f}")
```

<h1>INSIGHT 7</h1>

```python
plt.figure(figsize=(10, 6))
plt.scatter(internet['Internet users'], internet['Birth rate'], color='orange')
plt.title('Birth Rate vs. Internet Users')
plt.xlabel('Internet Users (%)')
plt.ylabel('Birth Rate')
plt.grid()
plt.show()
```

<h1>INSIGHT 8</h1>

```python
high_birth_rate_list = internet[internet['Birth rate'] > 40][['Country Name', 'Birth rate']]
print("Countries with Birth Rates Over 40:")
print(high_birth_rate_list)
```

<h1>INSIGHT 9</h1>

```python
plt.figure(figsize=(10, 6))
income_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])
plt.title('Income Group Distribution')
plt.ylabel('')  
plt.show()
```

<h1>INSIGHT 10</h1>

```python
correlation_value = internet['Birth rate'].corr(internet['Internet users'])
print(f"Correlation Between Birth Rate and Internet Users: {correlation_value:.2f}")
```
